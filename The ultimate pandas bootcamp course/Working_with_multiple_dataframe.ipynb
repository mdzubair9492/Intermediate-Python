**Inner vs Outter joins**

```python
#pd.merge()
# how='inner' -> 1. only the common keys are selected 2. similar to set intersection
# how='outer' -> 1. all keys are selected 2. similar to set union
```

```python
pd.merge(ivies,regions,how='inner')
```

```python
pd.merge(ivies,regions,how='outer')
```

**left vs right joins**

```python
pd.merge(ivies,regions,how='left') # we want to preserve the key indices of left object (ivies),merge data and discard everything else (set a - set b)
```

```python
ivies
```

```python
pd.merge(ivies,regions,how='right') # we want to preserve the right indices (regions),merge data and discard everything else ( set b - set a)
```

```python
regions
```
