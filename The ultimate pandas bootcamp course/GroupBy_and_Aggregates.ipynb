**Groupby() transformations**

```python
# groupby() + transform() for inplace transformation applied at subgroup level
```

```python
# Q) convert raw global sales to within genre standard scores
```

```python
games_relative=games.loc[:,['Name','Genre','Platform','Global_Sales']]
```

```python
games_relative.set_index(['Name','Platform']).groupby('Genre').transform(lambda x: (x - x.mean()) / x.std())\
        .sort_values(by='Global_Sales',ascending=False)
```

```python
# so grand theft auto v is highest seller relative to genre than the other
```

```python
#  groupby() +...

#  .agg() -> typically reduces the dimensions of the dataset
#  .filter() -> include / exclude without changing the dimensions
#  .transform() -> change the values in place without altering the shape
```

**There also apply()**

```python
ps3= games.loc[games.Platform=='PS3',['Name','Genre','EU_Sales','Global_Sales']]
```

```python
ps3
```

```python
# the paradigm : subgroup -> apply() -> output
```

```python
ps3.groupby('Genre').apply(lambda x:'solid' if x.EU_Sales.sum() >50 else 'weak')
```

```python
def sales_detail(sg):
  level='solid' if sg.EU_Sales.sum() >50 else 'weak'
  variability='volatile' if sg.EU_Sales.std()/sg.EU_Sales.mean() >2 else 'steady'
  return (variability,level+' sales')
```

```python
ps3.groupby('Genre').apply(sales_detail)
```

```python
"""
The coefficient of variation measures the ratio of the standard deviation to the mean. The standard deviation is used more often when we want to measure
the spread of values in a single dataset. The coefficient of variation is used more often
when we want to compare the variation between two different datasets.

https://www.statology.org/coefficient-of-variation-vs-standard-deviation/#:~:text=The%20coefficient%20of%20variation%20measures,variation%20between%20two%20different%20datasets.
"""
```

```python
ps3.groupby('Genre').apply(lambda x:print(x.info(),'\n'))
```
