**SKILL CHALLENGE**

*1. Starting with the main dataframe (sat) , create a pivot table that summarizes the total student enrollement (Student Enrollment) across all 5 borough. which boroughs have the high and low student bodies*

```python
sat.pivot_table(index='Borough',values='Student Enrollment',aggfunc='sum').sort_values(by='Student Enrollment',ascending=False)
```

```python
# or
```

```python
sat.pivot_table(index='Borough',values='Student Enrollment',aggfunc='sum').nlargest(5,columns='Student Enrollment')
```

```python
sat.pivot_table(index='Borough',values='Student Enrollment',aggfunc='sum').idxmin()
```

```python
sat.pivot_table(index='Borough',values='Student Enrollment',aggfunc='sum').idxmax()
```

*2. Modify the pivot table from the step above to also reflect the average student body size(Student Enrollment) across boroughs, in the same pivot table*

```python
sat.pivot_table(index='Borough',values='Student Enrollment',aggfunc=['sum','mean']).rename(axis=1,level=0,mapper={'sum':'Total student','mean':'Avg student body size'})
```
