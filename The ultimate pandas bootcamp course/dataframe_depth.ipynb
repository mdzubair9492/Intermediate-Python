*1. create a standalone function that,

3.   'king of popular' if x is between 220 and 600 (non-inclusive)
4.   'popular' if x is between 600 and 200 (non-inclusive)

```python
def standalone(x):
  if x<220:
    return 'relatively unknows'
  elif(x < 600):
    return 'king of popular'
  elif(x < 2000):
    return 'popular'
  else:
    return 'super-popular'
```
